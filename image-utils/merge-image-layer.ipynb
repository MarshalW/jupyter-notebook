# 图层的覆盖

```python
from PIL import Image
from IPython import display
import IPython

def overlay_images(background_path, overlay_path, output_path, position=(0, 0), transparency=0):
    # 打开背景图和叠加图
    background = Image.open(background_path)
    overlay = Image.open(overlay_path)

    # 设置叠加图的透明度
    overlay = overlay.convert("RGBA")
    overlay_with_transparency = Image.new("RGBA", overlay.size)
    for x in range(overlay.width):
        for y in range(overlay.height):
            r, g, b, a = overlay.getpixel((x, y))
            overlay_with_transparency.putpixel((x, y), (r, g, b, int(a * transparency)))

    # 将叠加图叠加在背景图上
    background.paste(overlay_with_transparency, position, overlay_with_transparency)

    # 保存结果
    background.save(output_path)

overlay_images("test1.jpg", "test2.png", "output.png", transparency=1)

display.display(display.Image(filename="output.png"))
```
